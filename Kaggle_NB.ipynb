```python
import pandas as pd
dataset = pd.read_csv("/content/heart.csv")
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
upd=['Sex','ChestPainType','RestingECG','ExerciseAngina','ST_Slope']
dataset[upd]=dataset[upd].apply(LabelEncoder().fit_transform)
```

```python
X = dataset.drop(['HeartDisease'], axis=1)
y = dataset['HeartDisease']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB().fit(X_train,y_train)
clf.predict(X_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f} %".format(acc*100))
from sklearn.metrics import precision_score
pre=precision_score(y_test, y_pred, average='weighted')
print("Precision: {:.2f} %".format(pre*100))
from sklearn.metrics import recall_score
re=recall_score(y_test, y_pred, average='macro')
print("Recall: {:.2f} %".format(re*100))
```

```python
from sklearn.naive_bayes import BernoulliNB
clf = BernoulliNB(binarize=2.5).fit(X_train,y_train)
clf.predict(X_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f} %".format(acc*100))
from sklearn.metrics import precision_score
pre=precision_score(y_test, y_pred, average='weighted')
print("Precision: {:.2f} %".format(pre*100))
from sklearn.metrics import recall_score
re=recall_score(y_test, y_pred, average='macro')
print("Recall: {:.2f} %".format(re*100))
```

```python
import pandas as pd
dataset = pd.read_csv("/content/HepatitisCdata.csv")
dataset.drop('Unnamed: 0',axis='columns', inplace=True)
dataset.head()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
dataset["Sex"]=labelencoder.fit_transform(dataset["Sex"])
dataset.tail()
```

```python
column_means = dataset.mean()
dataset= dataset.fillna(column_means)
dataset.tail()
```

```python
X = dataset.drop(['Category'], axis=1)
y = dataset['Category']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
clf = GaussianNB().fit(X_train,y_train)
clf.predict(X_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f} %".format(acc*100))
from sklearn.metrics import precision_score
pre=precision_score(y_test, y_pred, average='weighted')
print("Precision: {:.2f} %".format(pre*100))
from sklearn.metrics import recall_score
re=recall_score(y_test, y_pred, average='macro')
print("Recall: {:.2f} %".format(re*100))
```

```python
from sklearn.naive_bayes import MultinomialNB
clf = MultinomialNB().fit(X_train,y_train)
clf.predict(X_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f} %".format(acc*100))
from sklearn.metrics import precision_score
pre=precision_score(y_test, y_pred, average='weighted')
print("Precision: {:.2f} %".format(pre*100))
from sklearn.metrics import recall_score
re=recall_score(y_test, y_pred, average='macro')
print("Recall: {:.2f} %".format(re*100))
```

```python
from sklearn.naive_bayes import BernoulliNB
clf = BernoulliNB(binarize=2.5).fit(X_train,y_train)
clf.predict(X_train)
y_pred = clf.predict(X_test)
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test, y_pred)
print("Accuracy: {:.2f} %".format(acc*100))
from sklearn.metrics import precision_score
pre=precision_score(y_test, y_pred, average='weighted',zero_division=1)
print("Precision: {:.2f} %".format(pre*100))
from sklearn.metrics import recall_score
re=recall_score(y_test, y_pred, average='macro')
print("Recall: {:.2f} %".format(re*100))
```
